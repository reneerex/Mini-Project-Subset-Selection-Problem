```python
import random as r
S={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
S_list = list(S)
result = []
for i in range(1000):
  x = r.choices(S_list, k=5)
  if sum(x) == 0:
    result.append(x)
print(result)
```

```python
import random as r
S={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
S_list = list(S)
result = []
for i in range(1000):
  k = r.randint(3,7)
  x = r.choices(S_list, k=k)
  if sum(x) == 0:
    result.append(x)
print(result)
```
